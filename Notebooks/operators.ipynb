```python
# write a python program to check whether a triangle is euquilateral ,scalen , isoceles
# first check triangle is valid or invalid
a=int(input("enter a number"))
b=int(input("enter a number"))
c=int(input("enter a number"))
if (a+b>c ) and (b+c>a) and (a + c >b):
    print("valid triangle")
    if a == b == c:
        print("equilateral triangle") # all sides equal
    elif a==b or b==c or c==a:
        print(" isosceles triangle") # any two side equal
    else:
        print("scalene") # all sides different
else:
    print("invalid")


```

```python
# write a python program to print sum of first n even and odd number

n = int(input("enter a number"))
sum_even=0
sum_odd=0
for i in range(1,(2*n)+1,1):
    if i%2==0:
        sum_even=sum_even+i
    else:
        sum_odd+=i
print("sum of even number : ",sum_even)
print("sum of odd number :",sum_odd)
```
