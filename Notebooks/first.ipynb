```python
def prime():
    n=int(input("enter a number"))
    if n>1 and all( n %i!=0 for i in range(2,n)):
        print("prime")
    else:
        print("not a prime")
prime()
```

```python
#with the help of function to print a perfect number or not

def is_perfect(num):
    ans=0
    for i in range(1,num):
        if num%i==0:
            ans+=i
    if ans==num:
        return "perfect number"
    else:
        return"not a perfect number"
a=int(input("enter a number"))
print(is_perfect(a))




```
