```python
# dict(variable with values)
d=dict(name=["abc","pqr","xyz"],subject=["da","ds","de"])
print(d)
```

```python
#dict comprehension
# concise way to create a new dictionary
# it is used to solve smaller probelm in a program
# syntax-- {k:v looping statement conditional statement}

#{'name': ['abc', 'pqr', 'xyz'], 'subject': ['da', 'ds', 'de']}
d={}
for i in range (1,6,1):
    d[chr(96+i)]=i**3
print(d)
```

```python
# set 
# collection of element
# denoted by {} with csv
# hetrogeneous in nature ,unordered,unindexed,mutuable
# stored only unique value element.
s=set() # set constructor
print(s)
```

```python
s=set((10,20,30,"hello",40,20,20,10,10.5))
print(s)
```

```python
s={10,20,30,"hello",10.5,10,20,10}
print(s)
print(type(s))
# print(s[0]) #unindexed
s.add(100)   # mutuable
print(s)
```

```python
# direct iteration
s={10,20,30,"hello",10.5,10,20,10}
for i in s:
    print(i)
```
