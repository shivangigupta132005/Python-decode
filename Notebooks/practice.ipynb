```python
# 
start=int(input("enter a number"))
end=int(input("enter a number"))
sum_even=0
count_even=0
sum_odd=0
count_odd=0
for i in range(start,end+1):
    if i%2==0:
        sum_even=sum_even+i
        count_even=count_even+1
    else:
        sum_odd+=i
        count_odd+=1
print("sum of even",sum_even)
print("count of even",count_even)
print("sum of odd",sum_odd)
print("count of odd",count_odd)
    


```

```python
#fibonacci series
a=0
b=1
n=int(input("enter a number"))
print(a,b,end=" ")
for i in range(2,n):
    c=a+b
    print(c,end=" ")
    a=b
    b=c
    
```

```python
n=int(input("enter a number"))
count=0
for i in range(1,n+1):
    if n%i==0:
        count=count+1
if count==2:
    print("prime number")
else:
    print("not a prime number")
            
```

```python
n=int(input("enter a number"))
count=0
i=1
while(i<=n):
    if i%n==0:
        count=count+1
if count==2:
    print("prime number")
else:
    print("composite number")
```

```python
#reverse of a number
n=int(input("enter any number :"))
dig=0
while (n!=0):
    digit=n%10
    dig=(dig*10)+digit
    n=n//10
print(dig)
```
