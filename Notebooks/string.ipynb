```python
# write a python program to print no of vowela AND no of consonant and no of spaces
s="welcome to city of lakes"
count_v=0
count_c=0
count_s=0
for i in s:
    if i in ("aeiou"):
        count_v+=1
    
    elif i==" ":
        count_s+=1
    else:
        count_c+=1
print(count_v)
print(count_c)
print(count_s) 
```

```python
# count() - that return occurence of character in a given string.
# strobj.count(character,[start],[end])
s="welcome to city of lakes"
print(s.count("o"))
print(s.count("e"))
print(s.count("e",2,22))
print(s.count("e",2,5))
print(s.count("z"))
```

```python
# index()  return position of any character/substring.
# strobj.index(character/substring,[start],[end])
s="welcome to bhopal"
print(s.index("o"))
print(s.index("o",5))
print(s.index("bhopal"))
print(s.index("z"))
```

```python
# find()  return position of any character/substring.
# strobj.find(character/substring,[start],[end])
s="welcome to bhopal"
print(s.find("o"))
print(s.find("o",5))
print(s.find("bhopal"))
print(s.find("z"))
```
