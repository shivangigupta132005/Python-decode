```python
import time
size=100000
list1=range(size)
list2=range(size)
start = time.time()
result = [x+y for x,y in zip(list1,list2)]
end = time.time()
print("python list time:" ,end-start)

array1 = np.arange(size)
array2 = np.arange(size)
start= time.time()
result = array1 + array2
end = time.time()
print("python array time",end - start)
```
