```python
num=1000
even_num=0
odd_num=0

while (num<=2000):
    if num%2==0:
        even_num+=1
    else:
        odd_num+=1
    num+=1
print("count of even number is :",even_num)
print(" count of odd number is :",odd_num)
```

```python
# write a python program to print reverse of any given number

num=int(input("enter any number"))
rev=0
while num!=0:
    dig=num%10
    rev=rev*10+dig
    num=num//10
print("the reverse of number is ",rev)
    
```

```python
# write a python to check whether a given number is an amstrong number or not

n=int(input("enter a number"))
a=len(str(n))
sm=0
old=n

while n!=0:
    dig=n%10
    sm=sm+(dig**a)
    n=n//10
if sm==old:
    print(old,"amstrong")
else:
    print("not amstrong")
```

```python
# write a python to to check is pallindrom or not
n=int(input("enter a number"))
rev=0
number=n
while n!=0:
    dig=n%10
    rev= rev*10+dig
    n=n//10
if number==rev:
    print(rev,"pallindrom")
else:
    print("not a pallindrom")
```
