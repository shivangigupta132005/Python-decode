```python
# 
l=[1,4,3,2,5,1,5,7,7,7,3,2,9,8,7,7]
d={}
for i in l:
   if i in d:
       d[i]+=1
   else:
       d[i]=1

max_value=max(d.values())
for i,j in d.items():
    if j==max_value:
        print(i)

    
```

```python
l=[1,4,3,2,5,1,5,7,7,7,3,2,9,8,7,7]
count=0
sum=0
for i in l:
    count+=1
    sum+=i
    mean=sum/count
print("mean is :",mean)
sorted_list=(sorted(l))
#print(sorted_list)
n=len(sorted_list)
if n%2!=0:
    median=sorted_list[n//2]
else:
    median=(sorted_list[n//2]+sorted_list[n//2 -1])/2
print("median is :",median)
d={}
for i in l:
    if i in d:
        d[i]+=1
    else:
        d[i]=1
max_value=max(d.values())
for i,j in d.items():
    if j==max_value:
        print("mode is :",i)
```

```python
# union |
# intersection &
# difference -
# symmetric_difference ^

s1={1,2,3,5,7,9}
s2={11,12,16,7,3}

print(s1.union(s2))
print(s1.intersection(s2))
print(s1.difference(s2))
print(s1.symmetric_difference(s2))
```
