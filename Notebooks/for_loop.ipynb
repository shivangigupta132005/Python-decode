```python
# wap to print first n even and n odd number?

n = int(input("enter a number"))
for i in range(1,(2*n)+1,1):
    if i%2==0:
        print(i,"even")
    else:
        print(i,"odd")
```

```python
# wap to print the sum of n natural number ?
n=int(input("enter a number"))
sum=0
for i in range(1,n+1,1):
    sum=sum+i
    print(i)
print(sum)
```

```python
# write a python program to print the sum of even and odd number ....? n- user input



n=int(input("enter a number"))
even_sum=0
odd_sum=0
for i in range(1,(2*n)+1,1):
    if i%2==0:
        even_sum+=i
        print(i,"even")
    else:
        odd_sum+=i
        print(i,"odd")
print("sum of even number :",even_sum)
print("sum of odd number :",odd_sum)
        
```
