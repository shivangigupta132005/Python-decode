```python

#prime number
n=int(input())
isprime=True
if n<=1:
    print("not a prime")
else:
    for i in range (2,n):
        if n%i==0:
            
            isprime=False
            print("not a prime")
            break
    if isprime:
        print("prime")
    
```

```python
#fibonacci series
n=int(input("enter a number"))
fi=0
se=1


for i in range(n):
    print(fi,end=" ")
    nx=fi+se
    fi=se
    se=nx
    print(nx,end=" ")
```
