```python
#STACK
l=[]
while True:
    ch=int(input('''
    1.push
    2.pop
    3.peek
    4.display
    5.exit 
    '''))
    
    if ch==1:
        n=int(input("enter any number"))
        l.append(n)
        print(l)
    elif ch==2:

        if len(l)==0:
            print("empty stack :",l)
        else:
            l.pop()
            print(l)
    elif ch==3:
        if len(l)==0:
            print("empty stack :",l)
        else:
            print("last element :",l[-1])
    elif ch==4:
        print("stack element :",l)
    elif ch==5:
        break
    else:
         print("invalid stack")



```

```python
# Queue
l=[]
while True:
    ch=int(input('''
    1.enqueue
    2.dequeue
    3.front element
    4.rear element
    5.display
    6.exit
    '''))
    
    if ch==1:
        n=int(input("enter any number"))
        l.append(n)
        print(l)
    elif ch==2:

        if len(l)==0:
            print("empty queue :",l)
        else:
            l.pop(0)
            print(l)
    elif ch==3:
        if len(l)==0:
            print("empty queue :",l)
        else:
            print("last element :",l[0])
    elif ch==4:
        if len(l)==0:
            print("empty queue :",l)
        else:
            print("last element :",l[-1])

    elif ch==5:
        print("queue is : ",l)
    elif ch==6:
        break
    else:
         print("invalid stack")
```

```python
l=["madam","bhopal","aa","decode","Naman"]
l1=[]
for i in l:
    rev=""
    for j in i:
        rev=j+rev
    if i==rev:
        l1.append(i)
print(l1)
```

```python
l=["madam","bhopal","aa","decode","Naman"]
l1=[]
for i in l:
    if i==i[::-1]:
        l1.append(i)
print(l1)
print([i for i in l if i==i[::-1]])
```
