```python
# stack
l=[]
while True:
    ch=int(input('''
        1.push
        2.pop
        3.peek
        4.display
        5.exit
        '''))
    if ch==1:
        n=input("enter any number: ")
        l.append(n)
        print(l)
    elif ch==2:
        if len(l)==0:
            print("empty stack")
        else:
            p=l.pop()
            print(p)
            print(l)
    elif ch==3:
        if len(l)==0:
            print("empty stack")
        else:
            print("last element of stack: ",l[-1])
    elif ch==4:
        print("display stack: ",l)
    elif ch==5:
        break
    else:
        print("invalid input")
```

```python
# queue
l=[]
while True:
    ch=int(input('''
        1.enqueue
        2.dequeue
        3.front element
        4.rear element
        5.display
        6.exit
        '''))
    if ch==1:
        n=input("enter any number: ")
        l.append(n)
        print(l)
    elif ch==2:
        if len(l)==0:
            print("empty queue")
        else:
            del l[0]
            print(l)
    elif ch==3:
        if len(l)==0:
            print("empty queue")
        else:
            print("front element: ",l[0])
    elif ch==4:
        if len(l)==0:
            print("empty queue")
        else:
            print("rear element: ",l[-1])
    elif ch==5:
        print("display queue: ",l)
    elif ch==6:
        break
    else:
        print("invalid input")
```
