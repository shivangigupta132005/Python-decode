```python
# Write a program to calculate the sum of series up to n term. For
# example, if n =5 the series will become 5 + 55 + 555 + 5555+ 55555
# Output: 61725
n=int(input("enter a number"))
sum=0
sum_digit=0
for i in range(1,n+1):
    sum=sum*10+n
    sum_digit+=sum
print(sum_digit)

```
